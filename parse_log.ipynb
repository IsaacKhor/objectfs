```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
times = {}
df = {"timestamp": [], "time": []}
lines = open('/mnt/ramdisk/log_deferred_commit.txt', 'r').readlines()
for line in lines:
    line = line.split("|")
    t = line[0]
    #print(line)
    time = float(line[1].split(",")[0].split(":")[1])
    if t not in times.keys():
        times[t] = [time]
    else:
        times[t].append(time)
    df["timestamp"].append(t)
    df["time"].append(time)
print([(key, sum(times[key])) for key in times.keys()])
print([(key, len(times[key])) for key in times.keys()])
draw(df)
#detached thread
```

```python
def draw(df):
    sns.set(style="ticks")

    # Initialize the figure with a logarithmic x axis
    f, ax = plt.subplots(figsize=(7, 6))
    ax.set_xscale("log")

    # Load the example planets dataset
    df = pd.DataFrame.from_dict(df)

    # Plot the orbital period with horizontal boxes
    sns.boxplot(x="time", y="timestamp", data=df,
                whis=[0, 100], width=.6, palette="vlag")

    # Add in points to show each observation
    sns.stripplot(x="time", y="timestamp", data=df,
                size=4, color=".3", linewidth=0)

    # Tweak the visual presentation
    ax.xaxis.grid(True)
    ax.set(ylabel="")
    sns.despine(trim=True, left=True)
```
